```python
TRAIN_PROCESSED_FILE = 'train-processed.csv'
TEST_PROCESSED_FILE = 'test-processed.csv'
POSITIVE_WORDS_FILE = 'positive-words.txt'
NEGATIVE_WORDS_FILE = 'negative-words.txt'
```

```python
file = open('positive-words.txt', 'r')
```

```python
p=[]
for line in file:
    #print(line)
    p.append(str(line).split('\n')[0])
```

```python
file = open('negative-words.txt', 'r')
```

```python
q=[]
for line in file:
    #print(line)
    q.append(str(line).split('\n')[0])
```

```python
file = open('test-processed.csv', 'r')
```

```python
import pandas as pd
```

```python
predictions = []
for line in file:
    tweet_id, tweet = line.strip().split(',')
    pos_count, neg_count = 0, 0
    for word in tweet.split():
        if word in p:
            #print("positive word found :",word)
            pos_count += 1
        elif word in q:
            #print("negative word found :",word)
            neg_count += 1
    #print(pos_count, neg_count)
    prediction = 1 if pos_count >= neg_count else 0
    predictions.append((tweet_id,tweet,prediction))
```

```python
with open('output_senti.csv', 'w') as csv:
        csv.write('id,tweet,prediction\n')
        for tweet_id,tweet,pred in predictions:
            csv.write(tweet_id)
            csv.write(',')
            csv.write(tweet)
            csv.write(',')
            csv.write(str(pred))
            csv.write('\n')
```

```python
file = open('train-processed.csv', 'r')
```

```python
predictions_train = []
for line in file:
    tweet_id,pred,tweet = line.strip().split(',')
    pos_count, neg_count = 0, 0
    for word in tweet.split():
        if word in p:
            #print("positive word found :",word)
            pos_count += 1
        elif word in q:
            #print("negative word found :",word)
            neg_count += 1
    #print(pos_count, neg_count)
    prediction = 1 if pos_count > neg_count else 0
    predictions_train.append((tweet_id,tweet,prediction, pred))
```

```python
with open('output_senti_train.csv', 'w') as csv:
        csv.write('id,tweet,prediction,pred\n')
        for tweet_id,tweet,prediction, pred in predictions_train:
            csv.write(tweet_id)
            csv.write(',')
            csv.write(tweet)
            csv.write(',')
            csv.write(str(prediction))
            csv.write(',')
            csv.write(str(pred))
            csv.write('\n')
            
```
