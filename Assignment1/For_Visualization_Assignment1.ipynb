```python
import pandas as pd
import random
import matplotlib.pyplot as plt
```

```python
filename = "master_csv.csv"
n = sum(1 for line in open(filename)) - 1 #number of records in file (excludes header)
s = 100000 #desired sample size
skip = sorted(random.sample(range(1,n+1),n-s)) #the 0-indexed header will not be included in the skip list
df2= pd.read_csv(filename, skiprows=skip)
```

```python
#Yearly Traffic
fig = plt.figure(figsize=[12,9])
x=df2['date'].apply(lambda x : x.split('-')[1]).value_counts().sort_index().reset_index()['index']
y=df2['date'].apply(lambda x : x.split('-')[1]).value_counts().sort_index().reset_index()['date']
ax1 = fig.add_subplot(111)
ax1.set_xlabel('Months')
ax1.set_ylabel('Number of Hits')
ax1.set_title('Yearly Traffic')
ax1.plot(x,y)
plt.show()
```

```python
#Hourly Traffic
fig = plt.figure(figsize=[12,9])
x=df2['time'].apply(lambda x : x.split(':')[0]).value_counts().sort_index().reset_index()['index']
y=df2['time'].apply(lambda x : x.split(':')[0]).value_counts().sort_index().reset_index()['time']
y=y/len(df2['date'].value_counts())
ax1 = fig.add_subplot(111)
ax1.set_xlabel('Hour of Day')
ax1.set_ylabel('Avg Number of Hits/ Day')
ax1.set_title('Hourly Traffic')
ax1.plot(x,y)
plt.show()
```

```python
#Average Number of Hits per day VS Hour of Day
fig = plt.figure(figsize=[12,9])
x=df2['time'].apply(lambda x : x.split(':')[0]).value_counts().sort_index().reset_index()['index']
y=df2['time'].apply(lambda x : x.split(':')[0]).value_counts().sort_index().reset_index()['time']
y=y/len(df2['date'].value_counts())
ax1 = fig.add_subplot(111)
ax1.set_xlabel('Hour of Day')
ax1.set_ylabel('Avg Number of Hits/ Day')
ax1.set_title('Hourly Traffic')
ax1.plot(x,y)
plt.show()
```
